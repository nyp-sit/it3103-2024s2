## What are Agents?

You might have seen the 2008 movie _Iron Man_, in which Tony Stark created an A.I. software called [J.A.R.V.I.S.](https://en.wikipedia.org/wiki/J.A.R.V.I.S.) which operated his Iron Man armour suit. He could have conversations with J.A.R.V.I.S., give it instructions, and have it carry out those instructions autonomously. J.A.R.V.I.S. was his helpful, tireless butler that never sleeps and never takes a day off.

Today, LLM-powered autonomous agents have become a reality. It is still early days, so they are not exactly used to power superhero armour suits, but they can at least be used for question-and-answer systems in which the agent can autonomously search for answers to your queries, fetch the relevant information and arrive at an appropriate answer. The key element is that they can __make decisions on their own and carry out actions based on those decisions__.

## Today's LLM-based autonomous agents

The idea of agents is to use an LLM as a **reasoning engine** -- using it to take actions in a series of steps to accomplish a given goal, and make decisions about how to interact with the outside world. These interactions would involve **choosing appropriate tools** to use to extend the LLM's capabilities -- tools such as a search engine to look for information, or a calculator to do mathematical calculations, or a database to retrieve relevant data.

An LLM-based agent would typically have these components:

- An LLM (or more than one LLM) as a reasoning engine -- the "brain" behind its decision-making.
- Short-term memory, long-term memory or both, as a way to maintain state between parts of a multi-step chain of actions.
- Prompt templates to reuse commonly used prompts.
- Access to tools such as search engine, calculator, Python interpreter and many others.
- Access to external knowledge such as a database of some kind.

Together, these components work together as an agent system.

## Role of Agents in Enhancing LLM Interactions

Today's LLM apps like ChatGPT are known to be powerful text generators.

Whereas LLMs have already shown impressive abilities in natural language processing, giving **autonomous agent capability** and **tool-access to LLMs** will open the way to a wide range of **task automation** with many use cases. Agents will be able to handle the bulk of routine and tedious tasks, leaving humans more time to focus on more creative and higher level work.

This may likely reshape and disrupt some jobs, but in the long run will open up new opportunities for advancement.

## Types of Agents

Since the release of OpenAI's release of GPT-3 in 2020, there has been an explosion of research in the field of LLM-powered autonomous agents.

In the figure below from the research paper <a href="https://arxiv.org/abs/2308.11432">A Survey on Large Language Model based Autonomous Agents</a>, agents are broadly classified under 7 types:

- General Agents, e.g. ReAct, Auto-GPT, BabyAGI.
- Tool Agents, e.g. Toolformer, HuggingGPT and ToolBench.
- Simulation Agents, e.g. Generative Agent,  RecAgent, AgentSims.
- Embodied Agent, e.g. Inner Monologue, CO-LLM.
- Game Agent, e.g. DEPS, Voyager and GITM.
- Web Agent, e.g. WebShop, MIND2WEB.
- Assistant Agent, e.g. AgentGPT, ChatDev.

agents.png


Source: "A Survey on Large Language Model based Autonomous Agents" research paper by Wang et al.

**General Agents**

The general agents listed in this paper, such as [ReAct](https://arxiv.org/abs/2210.03629), [Auto-GPT](https://github.com/Significant-Gravitas/Auto-GPT) and [BabyAGI](https://github.com/yoheinakajima/babyagi), represent a general-purpose approach that allows LLMs to function as the 'brain' or **reasoning engine** in an agent system that can work autonomously to make decisions and take actions.

**Tool Agents**

Tool agents such as Toolformer and HuggingGPT are LLM-based autonomous agents which are able to use external tools and knowledge sources to enhance the LLM's capabilities. Such tool-using agents typically use an API (application programming interface) to access tools such as a search engine to perform information retrieval, a calculator to perform math, calendars to check dates, or even a Python console to run programming code.

**Simulation Agents**

Simulation agents are LLM-based agents used to conduct social experiments or research in place of actual humans. This offers a way to side-step ethical and financial issues if humans were to be used for such purposes. Such research has been done in the areas of community regulations, social recommendations, and the spread of harmful information in a community.

**Embodied Agents**

An embodied agent refers to an artificial intelligence (AI) system that exists within and interacts with a physical or virtual environment through a form of embodiment, such as a robot, virtual avatar, or other interactive interfaces. Unlike disembodied AI systems, which operate solely in software without a physical presence (e.g., chatbots or recommendation algorithms), embodied agents can perceive, act, and respond within their environment in a more tangible and interactive manner.

In the next section, we will look more closely at one implementation of the general agent -- **ReAct**, an approach to using reasoning and acting together as a system.

### ReAct vs Auto-GPT vs BabyAGI: A quick comparison

ReAct, Auto-GPT and BabyAGI are different approaches to LLM-powered autonomous agents. This is a fast-developing field and many other LLM agents are constantly appearing, building on existing ideas and experimenting with new ones.

- [ReAct](https://arxiv.org/abs/2210.03629) started life in October 2022 as a research paper, and has been implemented by [LangChain](https://python.langchain.com/docs/modules/agents/agent_types/react) and [LlamaIndex](https://docs.llamaindex.ai/en/stable/examples/chat_engine/chat_engine_react.html) -- two of the most popular LLM development frameworks. We will talk about ReAct below.

- [Auto-GPT](https://github.com/Significant-Gravitas/Auto-GPT) (a.k.a. AutoGPT) is an experimental open source Github project which was started on 23 March 2023. It is an attempt to make GPT-4 fully autonomous, although it also leverages GPT-3.5.

- [BabyAGI](https://github.com/yoheinakajima/babyagi) is developed by Yohei Nakajima and came out on 2 April 2023. It is built with OpenAI's GPT-4 and LangChain framework. It also uses Pinecone and Chroma vector databases.

Both Auto-GPT and BabyAGI can be given an objective, and they will break it down into sub tasks and execute them. Each is better at particular use cases. There are a lot of articles and videos about them if you search for them.

For the purpose of this training, we will focus on ReAct which is a more general approach with basic use cases.

# 3.2. ReAct: Reasoning and Acting

In October 2022, a group of researchers released the paper, [ReAct: Synergizing Reasoning and Acting in Language Models](https://arxiv.org/abs/2210.03629), in which they proposed to combine an LLM's ability to generate both reasoning traces and task-specific actions in an interleaved manner, to make it function effectively as an agent. The two parts that are combined or synergized are **reasoning** (e.g. chain-of-thought prompting) and **acting** (e.g. action plan generation), or **ReAct** for short.

By combining reasoning and acting, ReAct has been empirically evaluated to consistently outperform baselines that use only reasoning or acting alone across different domains.

An important feature of ReAct is that it generates reasoning traces -- as can be seen in the example below -- which can be inspected. This level of transparency enhances interpretability and trustworthiness.

## Overview of ReAct and How ReAct Enhances LLM Capabilities

Most people who have experienced the power of LLMs have also noticed certain limitations:

- LLMs have a knowledge cut-off, meaning it only knows information from the data it was trained on. It is not aware of anything new that has happened outside of its existing knowledge or after its training.

- Likewise, LLMs have no knowledge of proprietary information that is kept in the knowledge base of a company's archives.

- LLMs are not great at performing mathematical calculations. In fact, an ordinary calculator can do a more reliable job.

ReAct helps LLMs to overcome these inherent limitations.

Advantages of using ReAct:

- ReAct is more flexible and can recover from errors.
- ReAct can handle multi-step tasks.
- ReAct generates reasoning traces, which enhances interpretability and trustworthiness.

The agent will take multiple steps until it reaches the stopping condition.

## LangChain's ReAct agent

A wide range of tools accessible by LLMs have been made available in [LangChain](https://www.langchain.com/), a software framework used for developing applications using large language models.

The agent decides what to do based on the user input, and based on the results or observations it gets from its previous steps.

Below is an example of a series of reasoning and acting steps carried out by [LangChain's ReAct agent](https://python.langchain.com/docs/modules/agents/agent_types/react.html). Notice that the agent is forming its own thoughts, taking actions to use tools such as `search` and `calculator`, and based on the results of the actions, making follow-up decisions.

    Prompt: "Who is Leo DiCaprio's girlfriend? What is her current age raised to the 0.43 power?"

    > Entering new AgentExecutor chain...
    I need to find out who Leo DiCaprio's girlfriend is and then calculate her
    age raised to the 0.43 power.
    
    Action: Search
    Action Input: "Leo DiCaprio girlfriend"
    Observation: Camila Morrone
    
    Thought: I need to find out Camila Morrone's age
    Action: Search
    Action Input: "Camila Morrone age"
    Observation: 25 years

    Thought: I need to calculate 25 raised to the 0.43 power
    Action: Calculator
    Action Input: 25^0.43
    Observation: Answer: 3.991298452658078
    
    Thought: I now know the final answer
    Final Answer: Camila Morrone is Leo DiCaprio's girlfriend and her current

## Pro-Tip: Which LLM should you use for your ReAct agent?

As you now know, a ReAct agent leverages an LLM as the **reasoning engine** for the reasoning part of **reasoning and acting**.

For this purpose, a relatively powerful LLM with more parameters would work best for reasoning, as opposed to a smaller model with fewer parameters.

The most powerful LLMs today include:

- GPT-4o and o1, proprietary LLMs from OpenAI.
- PaLM 2, a proprietary LLM from Google.
- Llama 3, an open-source LLM released by Meta.
- Falcon, an open-source LLM released by TII.
- Command, a proprietary LLM from Cohere.
- Claude,  a proprietary LLM from Anthropic.

LLMs are constantly being updated and improved. What works best for a specific purpose this month may suddenly not be the best next month.

We recommend that you stay updated on the latest news about newly released LLMs and test them out for your own use case to determine what works best, bearing in mind to balance between cost and performance.

## Practical Applications of ReAct

The most practical use case of ReAct, as we have seen, is building a question-answering system, where the autonomous agent behaviour is coupled with tool use to augment the LLM's capability.

The authors of ReAct have noted that due to its flexible nature, ReAct works for a diverse range of tasks, including not just question-answering, but also fact verification, text games, and web navigation.

Developing with the [LangChain ReAct agent](https://python.langchain.com/docs/modules/agents/agent_types/react.html) means having the ability to handle tabular data from traditional SQL relational databases, which opens the way to integrate with legacy knowledge systems.

The technology behind autonomous agents is still very much in the early days, and it will continue to develop and mature. Possible future use cases might include personal digital assistants -- essentially a more capable version of Siri, which can write emails, make appointments, and help with other everyday tasks.

# 3.3. Retrieval Augmented Generation (RAG)

The key idea behind Retrieval Augmented Generation or RAG is that the performance of LLMs can be improved by augmenting the LLM's existing knowledge with more relevant knowledge that it can use for its completions.

The knowledge already stored in the parameters of an LLM, known as its **paramatric memory**, may not be enough to answer a user's query.

We can supplement the LLM's existing knowledge with new knowledge that we give to the LLM that is relevant to the query. This external knowledge is known as **non-parametric memory**.

With this extra relevant context, the LLM is then able to answer queries that are **knowledge-intensive**.

A knowledge-intensive query means it is asking for a response that is outside of the LLM's general knowledge. It requires knowledge that is more updated or proprietary, and boosts the LLM's ability to answer the query correctly.

## LLM's limitations addressed by RAG

LLMs have the following limitations:

- **Knowledge cut-off**

    An LLM is trained on data only up to a certain point. For example, GPT-3.5 is trained on data up to September 2021. It does not know about events that have happened after that cut-off date. If you ask who won the Singapore presidential election in 2023, it would not have the answer.

- **Lacking in domain-specific knowledge**, e.g. proprietary knowledge

    LLMs are trained on generally available data, generally from the Internet and publicly available sources. If you need proprietary information that is available only in the knowledge archive of a private organization, then the LLM would not be able to help you.

- **Hallucinations**

    An LLM is known to make things up if it does not have the needed information from its training data. Given an input prompt, what an LLM does is to generate the most statistically likely next word, and then the next most likely word after that, and so on. There is no built-in fact-checking capability. It can give you a confident-sounding answer which it plucked out of thin air but is completely false.

## Benefits of RAG

Retrieval Augmented Generation addresses the above limitations by providing:

- A framework for building LLM-powered systems that make use of external data sources to supplement the LLM's existing knowledge.

- An efficient way to to update the LLM's knowledge of the world. This avoids the expensive retraining and finetuning otherwise needed when you want to update the LLM with new data.

- The ability to "ground" the LLM on relevant facts as a way to significantly reduce hallucinations.

Thus, RAG offers these benefits:

- Reduced hallucinations.

- Access to new knowledge.

- External knowledge can be updated easily and efficiently.

- Reduced toxicity. Any toxic content found in an external knowledge source can be quickly removed.

- Improved transparency. Citations can be included in the LLM's completions, resulting in better explainability.

## How RAG Works: Combining Retrieval and Generation

Below is an example of how RAG works. The system combines the retrieved information with the input query and feeds this combination as a single prompt to the LLM. Thus the LLM has added context to answer the query, and it generates the correct reponse as a completion.

rag.png

**The Retriever component in RAG**

In RAG, the **Retriever** is the component that takes in the user's query and finds a piece of knowledge that can help the LLM answer the query accurately.

**What the Retriever does**

When the user inputs the query, it does not go to the LLM directly. Instead, it goes to the Retriever component in the RAG system.

In the Retriever, there is a **query encoder**, which takes the input query and encodes it into a vector, which is just a set of numbers that a computer can understand.

By comparing this vector with the external knowledge base -- which has also been encoded into vectors -- the Retriever can now perform a **similarity search** and find a piece of text which comes closest to answering the query.

Once this retrieved information is found, the RAG combines it with the input query and feeds the whole thing to the LLM. Now the LLM has the context it needs to answer the query.

The LLM is able to do this despite the fact that the Barbie movie was released recently in 2023, and therefore is not in its parametric knowledge which was trained before that.

Thus, RAG is able to augment the LLM with new knowledge that it needs.

### What is similarity search?

Similarity search is a technique that helps you find documents or pieces of text that are similar to a specific text you have in mind. This is how it works:

1. **Reference Text**: You start by selecting a piece of text that serves as your reference or target. This could be a sentence, a paragraph, or even a whole document. Let's say you have a reference text like, "The quick brown fox jumps over the lazy dog."

2. **Text-to-Features**: The similarity search algorithm takes this reference text, converts it into a set of features or characteristics, and expresses them in a list of numbers (vector) which a computer can understand. These features could be individual words, phrases, or even the structure of sentences. In our example, the features might include words like "quick," "brown," "fox," "jumps," "over," "the," "lazy," and "dog."

3. **Comparing Features**: Now, this algorithm goes through the knowledge collection of other texts, one by one, and analyzes their features in a similar way. It checks how closely the features of each text match the features of your reference text. This comparison can be based on things like word frequency, order of words, or more advanced techniques like word embeddings.

4. **Calculating Similarity**: After comparing all the texts, the algorithm assigns a similarity score to each one. The higher the score, the more similar that text is to your reference text. Texts that use similar words, phrases, or have a similar structure will get higher similarity scores.

5. **Presenting Results**: Finally, the algorithm provides you with a list of texts, ranked by their similarity scores. The texts at the top of the list are the most similar to your reference text, while those at the bottom are less similar.

Text similarity search helps you quickly find documents or pieces of text that share common themes, ideas, or language with your reference text. It's particularly valuable for tasks like content recommendation, where you want to suggest articles, products, or information that is closely related to what a user is interested in or searching for. It's also used in natural language processing tasks, sentiment analysis, and many other applications that involve working with text data.

Technically, there are various ways to measure **semantic similarity** (similarity in terms of meanings of words). The most commonly used metrics include:

- **Euclidean distance**. Intuitively, this is like the physical distance between 2 points in a vector space. The closer they are to each other, the more similar they are. Pythagoras theorem can be used to calculate this.

- **Cosine similarity**. This is calculated from the angle between 2 points, relative to an origin. The smaller the angle, the closer they are.

- **Dot product**. This is the same as cosine similarity, but also takes into account the distances of the 2 points from the origin. Dot product is useful because the distances can be used to encode an additional desirable feature other than similarity, such as the popularity of a Youtube video, so that these recommendations come up first before others.

````python
# Instantiate an LLM:
client = aoai(
    azure_endpoint=azure_endpoint,
    api_key=api_key,
    api_version="2023-05-15"
)

# Store a copy of the Barbie movie article pasted from Wikipedia:
text = """
Barbie is a 2023 American fantasy comedy film directed by Greta Gerwig from a screenplay she wrote with Noah Baumbach. Based on the eponymous fashion dolls by Mattel, it is the first live-action Barbie film after numerous computer-animated films and specials. The film stars Margot Robbie as the titular character and Ryan Gosling as Ken, and follows the pair on a journey of self-discovery following an existential crisis. It features a supporting cast that includes America Ferrera, Kate McKinnon, Issa Rae, Rhea Perlman, and Will Ferrell.

A live-action Barbie film was announced in September 2009 by Universal Pictures with Laurence Mark producing. Development began in April 2014, when Sony Pictures acquired the film rights. Following multiple writer and director changes and the casting of Amy Schumer and later Anne Hathaway as Barbie, the rights were transferred to Warner Bros Pictures in October 2018. Robbie was cast in 2019, after Gal Gadot turned down the role due to scheduling conflicts, and Gerwig was announced as director and co-writer with Baumbach in 2020. The rest of the cast was announced in early 2022, with principal photography occurring primarily at Warner Bros Studios, Leavesden, in England and at the Venice Beach Skatepark in Los Angeles from March to July of that year.

Barbie premiered at the Shrine Auditorium in Los Angeles on July 9, 2023, and was released in the United States on July 21, by Warner Bros Pictures. Its simultaneous release with Universal's Oppenheimer led to the "Barbenheimer" cultural phenomenon, which encouraged audiences to see both films as a double feature. The film received critical acclaim and has grossed over $1.404 billion, becoming the highest-grossing film of 2023 as well as the highest-grossing film by a solo female director, the highest-grossing film ever released by Warner Bros, and the 14th highest-grossing film of all time.

The idea of a live-action Barbie film had been in development at Cannon Films in the mid-1980s; little is known about it, other than a promotional blurb (with the plotline centering around Barbie showing her owner how all her dreams could come true), and it is unclear how far into development it got before Mattel ceased working with Cannon, following Cannon's live-action adaptation of Masters of the Universe (1987) flopping at the box-office.

Development on a film based on the Barbie toy line began in September 2009, when it was announced that Mattel had signed a partnership to develop the project with Universal Pictures and with Laurence Mark as producer, but nothing came to fruition. In April 2014, Mattel teamed with Sony Pictures to produce the film, which would have Jenny Bicks writing the screenplay and Laurie MacDonald and Walter F. Parkes producing through the Parkes+MacDonald Image Nation banner they created. Filming at the time was anticipated to begin by the end of the year. In March 2015, Diablo Cody was brought onto the project to rewrite the screenplay, and Amy Pascal joined the producing team. Sony Pictures would again have rewrites done to the screenplay later that year, hiring Lindsey Beer, Bert V. Royal, and Hillary Winston to write separate drafts.

In December 2016, Amy Schumer entered negotiations to star in the title role with Winston's screenplay; Schumer helped rewrite the script with her sister, Kim Caramele. In March 2017, Schumer exited negotiations, blaming scheduling conflicts with the planned June 2017 filming start; in 2023 she revealed she left the project due to creative differences with the film's producers at the time. That July, Anne Hathaway was under consideration for the title role, with Sony Pictures hiring Olivia Milch to rewrite the screenplay and approaching Alethea Jones to direct as a means to interest Hathaway into signing on. Jones was attached to direct by March 2018.

In August 2018, Robbie Brenner had been hired as producer by Mattel CEO Ynon Kreiz with the rights reverting to Mattel, with Kreiz having the intention to repossess the rights following the expiration of Sony Pictures's option. Later, Brenner was hired to run Mattel Films. The expiration of Sony Pictures' option on the project in October 2018 and its transfer to Warner Bros Pictures would see the departures of Hathaway, Jones, Macdonald, Parkes and Pascal. Margot Robbie would enter early talks for the role, with Patty Jenkins briefly considered for the director position. Kreiz was determined to cast Robbie in the titular role after meeting with her following his hiring as CEO as both he and Brenner felt that Robbie's appearance resembled that of a conventional Barbie doll and were impressed by her ideas. Initial meetings occurred at the Polo Lounge located in The Beverly Hills Hotel. Brenner eventually partnered with Robbie's production company, LuckyChap Entertainment, with Robbie's husband Tom Ackerley and Josey McNamara also being enlisted as producers. Robbie's casting was confirmed in July 2019.

In her capacity as a producer, Robbie pitched Barbie to Warner Bros herself. During the green-light meeting, she compared the film to Steven Spielberg's Jurassic Park (1993) and also jokingly suggested that it would gross over $1 billion. Later on, she approached Greta Gerwigâwhose previous films, particularly Little Women (2019), Robbie enjoyedâto be the screenwriter. Gerwig was in post-production for another film, and accepted the assignment on the condition that her partner, Noah Baumbach, also write the script. Gerwig would sign on to also direct the film in July 2021. Robbie said the film aimed at subverting expectations and giving audiences "the thing you didn't know you wanted". In August 2023, it was announced that Robbie would earn "roughly $50 million in salary and box office bonuses" as the star and producer of the film.
"""

# Do some chunking.
# Split the text into a list of sentences, by using full-stop as a delimiter.
# Store the list in a variable called texts.
texts = text.split('.')

# Loop through the sentences stored in `texts` and clean up by stripping out
# trailing spaces and new lines.
texts = [t.strip(' \n') for t in texts]

# Prepend the movie title to each sentence, to give it context for the LLM.
# Convert to numpy array.
texts = np.array([f"Barbie (film). {t}" for t in texts])

# Define the query encoder function:
def get_embedding(text_to_embed: str) -> list:
    """ Query encoder function, which takes a line of text as input
        and outputs a list of numbers which a computer can understand.

    Args:
        text_to_embed (str): A string of text which you want to turn into
            a vector.

    Returns:
        list: Embedding vector.
    """

    # Generate the embedding vector for a line of text:
    response = client.embeddings.create(
        input=text_to_embed,
        model="text-embedding-ada-002"
    )
    # Isolate the embedding vector which is a list of numbers:
    embedding = response.data[0].embedding

    # Output the embedding vector:
    return embedding

# Create a list of embedding vectors for each sentence of the knowledge base:
embeddings = [get_embedding(t) for t in texts]
# Note: embeddings is a list. The embedding for each sentence has a length of 1536.

# Instantiate Annoy search index.
# Note we are using Annoy's angular distance which is Euclidean distance of normalized vectors.
# The first argument len(embeddings[0]) is the length of each vector to be
# indexed, which is 1536.
search_index = AnnoyIndex(len(embeddings[0]), 'angular')

# Add all the knowledge base embedding vectors to the Annoy search index:
for i in range(len(embeddings)):
    search_index.add_item(i, embeddings[i])

search_index.build(10)         # Build 10 trees.
search_index.save('test.ann')  # Save to file. This will return True.
# You will be able to see the test.ann in the Colab files list.

# Define the search function to retrieve the most relevant sentences matching
# the user's query:
def search(query):
    """ Helper function to get nearest neighbours of the query text.

    Args:
        query (str): Query string.

    Returns:
        pd.DataFrame: DataFrame of the nth most similar texts ('texts'),
            and their corresponding distance ('distance') from the query.
    """

    # Generate the embedding vector for the query:
    query_embed = client.embeddings.create(
        input=query,
        model="text-embedding-ada-002"
    )

    # Isolate the query embedding vector which is a list of numbers:
    query_vector = query_embed.data[0].embedding

    # Use the Annoy search to retrieve the top 5 most relevant sentences,
    # a.k.a. the nearest neighbors:
    similar_item_ids = search_index.get_nns_by_vector(
        query_vector,
        n=5,  # Return the n nearest neighbors.
        include_distances=True
    )
    # Format the results:
    results = pd.DataFrame(data={'texts': texts[similar_item_ids[0]],
                                'distance': similar_item_ids[1]})

    return results

# # For debugging purpose:
# # Set display options to show long lines:
# pd.set_option('display.max_colwidth', None)
# query = "When was the Barbie movie released?"
# results = search(query)
# # Display the search results dataframe:
# results
````

Now that we have finished building the retriever, we can integrate it with the LLM so that we get an end-to-end pipeline, from user input to LLM output.

### Part 2: Integrate the retriever with the LLM

Run the following code cell to fit the pieces together.

````python
# Define a helper function to feed a prompt to the LLM and  get a response.
# This is the same function used last week.
def get_completion(
    prompt: str,
    model: str = "gpt-3.5-turbo",
    temperature: float = 0
) -> str:
    """ Helper function to get prompt completions.

    Args:
        prompt (str): The prompt for the LLM.
        model (str): The model name to be passed into the OpenAI API.
        temperature (int or float): Degree of randomness of LLM's output, from
            0 to 1. Default is 0.

    Returns:
        str: The completion from the LLM.
    """
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=deployment_name,
        messages=messages,
        temperature=temperature,
        #engine=deployment_name,
    )
    return response.choices[0].message.content

# Define end-to-end function to feed the user's input to the LLM.
def feed_llm_query(question):
    """ End-to-end function to feed the user's input to the LLM.
        This will construct a prompt for the LLM, incorporating the search
        results from the retriever.

    Args:
        question (str): Query which you want to ask the LLM.

    Returns:
        str: The completion from the LLM after giving it relevant info.
    """
    # Search the knowledge base:
    results = search(question)

    # Get the text of top results as a numpy array:
    context = np.array(results['texts'])

    # Prepare the prompt:
    prompt = (
        "Here is some relevant context from Wikipedia:\n\n"
        f"{context}\n\n"
        f"Question: {question}\n\n"
        "Extract the answer to the question from the context provided.\n"
        "If the context does not contain the answer, reply that the answer is not available."
    )
    print(prompt)

    return get_completion(prompt)
````
